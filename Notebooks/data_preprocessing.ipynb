```python
def  get_avg_word_len(x):
    words=x.split()
    words_len=0
    for word in words:
        words_len=words_len+len(word)
    return words_len/len(words)
    
```

```python
df['avg_word_len']=df['text'].apply(lambda x:get_avg_word_len(x))
```

```python
def cont_to_exp(x):
    if type(x) is str:
        for key in contractions:
            value=contractions[key]
            x=x.replace(key,value)
        return x
    else:
        return x
df['text']=df['text'].apply(lambda x: cont_to_exp(x))
```

```python
df['emails']=df['text'].apply(lambda x: re.findall(r'([a-zA-Z0-9._-]+@[a-zA-Z0-9._-]+\.[a-zA-Z0-9_-]+)', x))
df['emails_counts']=df['emails'].apply(lambda x: len(x) )
df['urls']=df['text'].apply(lambda x:re.findall(r'(https?://[^\s]+)',x))

df['urls_counts']=df['urls'].apply(lambda x: len(x))
df['text']=df['text'].apply(lambda x:re.sub('RT',"",x))
df['text']=df['text'].apply(lambda x: re.sub('[^A-Z a-z 0-9]+','',x))
df['text']=df['text'].apply(lambda x: " ".join(x.split())) 
```

```python
import unicodedata
def remove_accented_chars(x):
    x=unicodedata.normalize('NFKD',x).encode('ascii','ignore').decode('utf-8','ignore')
    return x
```

```python
df['text']=df['text'].apply(lambda x: remove_accented_chars(x))
```

```python

df['text']=df['text'].apply(lambda x:" ".join([t for t in x.split() if t  not in STOP_WORDS]))
```
